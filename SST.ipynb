```python
import pandas as pd
import numpy as np
from scipy import stats 
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import pycmap
import cmocean
```

```python
#Import and access data
#It's a large data set, so it takes a while to load :)
import requests
from netCDF4 import Dataset
import xarray as xr
def get_data(url):
    link = Dataset(url)
    return(xr.open_dataset(xr.backends.NetCDF4DataStore(link)))
sst_2013=get_data('https://www.ncei.noaa.gov/pub/data/cmb/ersst/v5/netcdf/ersst.v5.201301.nc#mode=bytes')
sst_2014= get_data('https://www.ncei.noaa.gov/pub/data/cmb/ersst/v5/netcdf/ersst.v5.201401.nc#mode=bytes')
sst_2015= get_data('https://www.ncei.noaa.gov/pub/data/cmb/ersst/v5/netcdf/ersst.v5.201501.nc#mode=bytes')
sst_2016= get_data('https://www.ncei.noaa.gov/pub/data/cmb/ersst/v5/netcdf/ersst.v5.201601.nc#mode=bytes')
sst_2017= get_data('https://www.ncei.noaa.gov/pub/data/cmb/ersst/v5/netcdf/ersst.v5.201701.nc#mode=bytes')
sst_2018= get_data('https://www.ncei.noaa.gov/pub/data/cmb/ersst/v5/netcdf/ersst.v5.201801.nc#mode=bytes')
sst_2019= get_data('https://www.ncei.noaa.gov/pub/data/cmb/ersst/v5/netcdf/ersst.v5.201901.nc#mode=bytes')
sst_2020= get_data('https://www.ncei.noaa.gov/pub/data/cmb/ersst/v5/netcdf/ersst.v5.202001.nc#mode=bytes')
sst_2021= get_data('https://www.ncei.noaa.gov/pub/data/cmb/ersst/v5/netcdf/ersst.v5.202101.nc#mode=bytes')
sst_2022= get_data('https://www.ncei.noaa.gov/pub/data/cmb/ersst/v5/netcdf/ersst.v5.202201.nc#mode=bytes')
sst_2023=get_data('https://www.ncei.noaa.gov/pub/data/cmb/ersst/v5/netcdf/ersst.v5.202301.nc#mode=bytes')
```

```python
#grab lon and lat values
lon_2013= sst_2013.lon.values
lat_2013= sst_2013.lat.values

lon_2014= sst_2014.lon.values
lat_2014= sst_2014.lat.values

lon_2015= sst_2015.lon.values
lat_2015= sst_2015.lat.values

lon_2016= sst_2016.lon.values
lat_2016= sst_2016.lat.values

lon_2017= sst_2017.lon.values
lat_2017= sst_2017.lat.values

lon_2018= sst_2018.lon.values
lat_2018= sst_2018.lat.values

lon_2019= sst_2019.lon.values
lat_2019= sst_2019.lat.values

lon_2020= sst_2020.lon.values
lat_2020= sst_2020.lat.values

lon_2021= sst_2021.lon.values
lat_2021= sst_2021.lat.values

lon_2022= sst_2022.lon.values
lat_2022= sst_2022.lat.values

lon_2023= sst_2023.lon.values
lat_2023= sst_2023.lat.values


#create 2D ssta datasets
sst1_2013= sst_2013['sst'].isel(lev=0, time=0)
sst1_2014= sst_2014['sst'].isel(lev=0, time= 0)
sst1_2015= sst_2015['sst'].isel(lev=0, time=0)
sst1_2016= sst_2016['sst'].isel(lev=0, time=0)
sst1_2017= sst_2017['sst'].isel(lev=0, time=0)
sst1_2018= sst_2018['sst'].isel(lev=0, time=0)
sst1_2019= sst_2019['sst'].isel(lev=0, time=0)
sst1_2020= sst_2020['sst'].isel(lev=0, time=0)
sst1_2021= sst_2021['sst'].isel(lev=0, time=0)
sst1_2022= sst_2022['sst'].isel(lev=0, time=0)
sst1_2023= sst_2023['sst'].isel(lev=0, time=0)
```

```python
#creat a figures
# Tried to make coloumns and rows, but the arrays wouldn't be on one figure
fig = plt.figure()
ax= plt.axes(projection =ccrs.Robinson())
ax.coastlines(resolution = '110m', color='black')
ax.add_feature(cfeature.LAND, color='lemonchiffon')
plt.pcolormesh(lon_2013, lat_2013, sst1_2013, cmap= cmocean.cm.haline, transform= ccrs.PlateCarree(),
               vmin= -3, vmax= 35)
plt.colorbar(label= 'Temperature (C)')
plt.title('2013 January Sea Surface Temperature')
ax.gridlines(crs= ccrs.PlateCarree(), color= 'white', linestyle= '--', alpha= 0.5 )
plt.savefig('2013_SST')


fig2= plt.figure()
axs= plt.axes(projection= ccrs.Robinson())
axs.coastlines(resolution= '110m', color= 'black')
axs.add_feature(cfeature.LAND, color='lemonchiffon')
plt.pcolormesh(lon_2014,lat_2014, sst1_2014, cmap= cmocean.cm.haline, transform= ccrs.PlateCarree(),vmin= -3, vmax= 35)
plt.colorbar(label= 'Temperature (C)')
plt.title('2014 January Sea Surface Temperature')
axs.gridlines(crs= ccrs.PlateCarree(), color= 'white', linestyle= '--', alpha= 0.5)
plt.savefig('2014_SST')

fig3= plt.figure()
axs3= plt.axes(projection= ccrs.Robinson())
axs3.coastlines(resolution= '110m', color= 'black')
axs3.add_feature(cfeature.LAND, color='lemonchiffon')
plt.pcolormesh(lon_2015,lat_2015, sst1_2015, cmap= cmocean.cm.haline, transform= ccrs.PlateCarree(),vmin= -3, vmax= 35)
plt.colorbar(label= 'Temperature (C)')
plt.title('2015 January Sea Surface Temperature')
axs3.gridlines(crs= ccrs.PlateCarree(), color= 'white', linestyle= '--', alpha= 0.5)
plt.savefig('2015_SST')

fig4= plt.figure()
ax4= plt.axes(projection= ccrs.Robinson())
ax4.coastlines(resolution= '110m', color= 'black')
ax4.add_feature(cfeature.LAND, color='lemonchiffon')
plt.pcolormesh(lon_2016,lat_2016, sst1_2016, cmap= cmocean.cm.haline, transform= ccrs.PlateCarree(),vmin= -3, vmax= 35)
plt.colorbar(label= 'Temperature (C)')
plt.title('2016 January Sea Surface Temperature')
ax4.gridlines(crs= ccrs.PlateCarree(), color= 'white', linestyle= '--', alpha= 0.5)
plt.savefig('2016_SST')

fig5= plt.figure()
ax5= plt.axes(projection= ccrs.Robinson())
ax5.coastlines(resolution= '110m', color= 'black')
ax5.add_feature(cfeature.LAND, color='lemonchiffon')
plt.pcolormesh(lon_2017,lat_2017, sst1_2017, cmap= cmocean.cm.haline, transform= ccrs.PlateCarree(),vmin= -3, vmax= 35)
plt.colorbar(label= 'Temperature (C)')
plt.title('2017 January Sea Surface Temperature')
ax5.gridlines(crs= ccrs.PlateCarree(), color= 'white', linestyle= '--', alpha= 0.5)
plt.savefig('2017_SST')

fig6= plt.figure()
ax6= plt.axes(projection= ccrs.Robinson())
ax6.coastlines(resolution= '110m', color= 'black')
ax6.add_feature(cfeature.LAND, color='lemonchiffon')
plt.pcolormesh(lon_2018,lat_2018, sst1_2018, cmap= cmocean.cm.haline, transform= ccrs.PlateCarree(),vmin= -3, vmax= 35)
plt.colorbar(label= 'Temperature (C)')
plt.title('2018 January Sea Surface Temperature')
ax6.gridlines(crs= ccrs.PlateCarree(), color= 'white', linestyle= '--', alpha= 0.5)
plt.savefig('2018_SST')

fig7= plt.figure()
ax7= plt.axes(projection= ccrs.Robinson())
ax7.coastlines(resolution= '110m', color= 'black')
ax7.add_feature(cfeature.LAND, color='lemonchiffon')
plt.pcolormesh(lon_2019,lat_2019, sst1_2019, cmap= cmocean.cm.haline, transform= ccrs.PlateCarree(),vmin= -3, vmax= 35)
plt.colorbar(label= 'Temperature (C)')
plt.title('2019 January Sea Surface Temperature')
ax7.gridlines(crs= ccrs.PlateCarree(), color= 'white', linestyle= '--', alpha= 0.5)
plt.savefig('2019_SST')

fig8= plt.figure()
ax8= plt.axes(projection= ccrs.Robinson())
ax8.coastlines(resolution= '110m', color= 'black')
ax8.add_feature(cfeature.LAND, color='lemonchiffon')
plt.pcolormesh(lon_2020,lat_2020, sst1_2020, cmap= cmocean.cm.haline, transform= ccrs.PlateCarree(),vmin= -3, vmax= 35)
plt.colorbar(label= 'Temperature (C)')
plt.title('2020 January Sea Surface Temperature')
ax8.gridlines(crs= ccrs.PlateCarree(), color= 'white', linestyle= '--', alpha= 0.5)
plt.savefig('2020_SST')

fig9= plt.figure()
ax9= plt.axes(projection= ccrs.Robinson())
ax9.coastlines(resolution= '110m', color= 'black')
ax9.add_feature(cfeature.LAND, color='lemonchiffon')
plt.pcolormesh(lon_2021,lat_2021, sst1_2021, cmap= cmocean.cm.haline, transform= ccrs.PlateCarree(),vmin= -3, vmax= 35)
plt.colorbar(label= 'Temperature (C)')
plt.title('2021 January Sea Surface Temperature')
ax9.gridlines(crs= ccrs.PlateCarree(), color= 'white', linestyle= '--', alpha= 0.5)
plt.savefig('2021_SST')

fig10= plt.figure()
ax10= plt.axes(projection= ccrs.Robinson())
ax10.coastlines(resolution= '110m', color= 'black')
ax10.add_feature(cfeature.LAND, color='lemonchiffon')
plt.pcolormesh(lon_2022,lat_2022, sst1_2022, cmap= cmocean.cm.haline, transform= ccrs.PlateCarree(),vmin= -3, vmax= 35)
plt.colorbar(label= 'Temperature (C)')
plt.title('2022 January Sea Surface Temperature')
ax10.gridlines(crs= ccrs.PlateCarree(), color= 'white', linestyle= '--', alpha= 0.5)
plt.savefig('2022_SST')

fig11= plt.figure()
ax11= plt.axes(projection= ccrs.Robinson())
ax11.coastlines(resolution= '110m', color= 'black')
ax11.add_feature(cfeature.LAND, color='lemonchiffon')
plt.pcolormesh(lon_2023,lat_2023, sst1_2023, cmap= cmocean.cm.haline, transform= ccrs.PlateCarree(),vmin= -3, vmax= 35)
plt.colorbar(label= 'Temperature (C)')
plt.title('2023 January Sea Surface Temperature')
ax11.gridlines(crs= ccrs.PlateCarree(), color= 'white', linestyle= '--', alpha= 0.5)
plt.savefig('2023_SST')

plt.rcParams.update({'font.size':12})
```
