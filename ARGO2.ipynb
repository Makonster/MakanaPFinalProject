```python
import pandas as pd
import numpy as np
from scipy import stats 
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import pycmap
import cmocean
```

```python
#Got help from Kathy for accesing the ARGO data
#And for setting up the query
#Due to large data acess only january was used
key = "d7672c4a-8277-4127-a1bf-45889999d607"
api = pycmap.API(token=key)

# access data by querying on a monthly basis 
query = ''' 
select datepart('year', time) year, datepart('month', time) month,
floor(lat/3)*3 lat, floor(lon/3)*3 lon, avg(depth) depth, avg(temp) temperature
from tblArgoCore_REP_Nov2023
where time between '2013-01-01' and '2013-01-31' 
and depth between 0 and 1000
group by 1, 2, 3, 4
order by 1, 2, 3, 4 
'''
# make query
data13=api.query(query)  
data13['temperature'].mean()
#Filter data
filtered_df13 = data13.query('505 >depth > 500')
fil13mean = filtered_df13.mean()

#Plot the data
#Tried to plot on one figure, but was given errors. Tried several work arounds to no avail. 
fig, ax = plt.subplots(figsize=(15,8))
ax.plot(filtered_df13['depth'], filtered_df13['temperature'], marker= "*", label= 'Temperature', markersize= 10,
        color= 'forestgreen', linestyle = 'None')
ax.set_xlabel('Temperature (C)', fontsize= '14')
ax.set_ylabel('Depth (M)', fontsize= '14')
ax.invert_yaxis()
ax.legend(loc= 'lower right')
ax.set_title('ARGO Float 2013', fontsize= '18')
ax.grid(linestyle= '-')
plt.savefig('2013_ARGO')
#used the prelectures and in class demos for set up
#Used Matplotlib website for color options and marker shape options
```
